```python
import audioread
import itertools
import numpy
import struct
import supriya
from pathlib import Path
from supriya.soundfiles import load
import matplotlib.axes
import matplotlib.pyplot as plt
```

```python
path = Path(supriya.__path__[0]) / "samples" / "sine_440hz_44100sr_16bit_octo.wav"
data, sample_rate = load(path)
array = numpy.array(data)
```

```python
fig, ax = plt.subplots(nrows=1)
fig.set_size_inches(8, 2)
fig.patch.set_alpha(1.0)
ax.set_facecolor("white")
ax.xaxis.label.set_color("white")
ax.tick_params(axis="x", colors="black")
ax.tick_params(axis="y", colors="black")
times = numpy.linspace(
    [0] * len(array), 
    [len(array[0]) / sample_rate] * len(array),
    num=len(array[0]),
)
plt.plot(times, array.transpose())
```
